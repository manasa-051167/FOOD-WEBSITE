```python
#Design and Implementation of deep neural network for Object detection

!pip install ultralytics
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load a pre-trained YOLOv8 model
model = YOLO("yolov8n.pt")  # 'n' version is lightweight; use 'yolov8s.pt' for better accuracy

# Function to detect objects in an image
def detect_objects(image_path):
    # Load image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for proper display

    # Run object detection
    results = model(image_path)

    # Plot the detection results
    results[0].show()  # Displays the image with bounding boxes

    # Save output image with bounding boxes
    output_path = "detected_objects.jpg"
    results[0].save(output_path)
    print(f"Detection results saved to {output_path}")
```

```python
image_path = '/content/drive/MyDrive/Dog.jpg' # Change this to your image file path
detect_objects(image_path)
```
